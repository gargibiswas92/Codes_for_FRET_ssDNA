```python
inp1 = open('new_dis_001.txt', 'r')
n1 = []
n2 = []
n3 = []
n4 = []
n5 = []
t = []
count = 0
for line in inp1:
    arr = line.split()
    n1.append(float(arr[1]))
    n2.append(float(arr[2]))
    n3.append(float(arr[3]))
    n4.append(float(arr[4]))
    n5.append(float(arr[5]))
    count = count + 1
    t.append(count)
    
import matplotlib.pyplot as plt
import matplotlib.pyplot 
plt.plot(t,n1, label='contact_1')
plt.plot(t,n2, label='contact_2')
plt.plot(t,n3, label='contact_3')
plt.plot(t,n4, label='contact_4')
plt.plot(t,n5, label='contact_5')

plt.xlabel('Time steps', fontsize=16)
plt.ylabel('Distance between ends ($\AA$)', fontsize=16)
plt.legend()
plt.title('Simulation1', fontsize=20)
plt.savefig('sim1.png', dpi=300, bbox_inches='tight')
```
